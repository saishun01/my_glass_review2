```python
from scipy import integrate

#1回目
wl1_arr = (df['wl1_12_ND64'].dropna()).values
wl2_arr = (df['wl2_12_ND8'].dropna()).values
wl3_arr = (df['wl3_1'].dropna()).values
wl4_arr = (df['wl4_12'].dropna()).values
wl5_arr = (df['wl5_12'].dropna()).values
wl6_arr = (df['wl6_1'].dropna()).values

H1_arr = (df['H1_1'].dropna()).values - BG_1
H2_arr = (df['H2_1'].dropna()).values - BG_2
H3_arr = (df['H3_1'].dropna()).values - BG_3
H4_arr = (df['H4_1'].dropna()).values - BG_4
H5_arr = (df['H5_1'].dropna()).values - BG_5
H6_arr = (df['H6_1'].dropna()).values - BG_6

I1 = integrate.simps(H1_arr, wl1_arr)
I2 = integrate.simps(H2_arr, wl2_arr)
I3 = integrate.simps(H3_arr, wl3_arr)
I4 = integrate.simps(H4_arr, wl4_arr)
I5 = integrate.simps(H5_arr, wl5_arr)
I6 = integrate.simps(H6_arr, wl6_arr)
```

```python
print(I1)
print(I2)
print(I3)
print(I4)
print(I5)
print(I6)
```

```python
N = np.arange(3,9)
I = [I1,I2,I3,I4,I5,I6]

plt.figure()
plt.scatter(N,I)
plt.yscale('log')
plt.xlabel(r'$n$')
plt.ylabel(r'Intensity (V$\cdot$nm)')
plt.xlim(2.8,8.2)
plt.ylim(10**(-4),10**(1))
plt.grid()
plt.show()
```
