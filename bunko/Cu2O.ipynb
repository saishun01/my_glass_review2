```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cu2O.csv')
df
```

```python
plt.figure()
ax = df.plot('wl_1', 'ref_1_ND64')
df.plot('wl_1', 'Cu2O_1_77K_ND4', ax=ax)
df.plot('wl_1', 'Cu2O_1_150K_ND4', ax=ax)
df.plot('wl_1', 'Cu2O_1_296K_ND4', ax=ax)
df.plot('wl_1', 'BG_1_ND4', ax=ax)
plt.show()
```

```python
df.mean()
```

```python
BG_1 = -0.048067
BG_23 = -0.046062
```

```python
df['T_1_77K'] = (df['Cu2O_1_77K_ND4']/0.23 - BG_1/0.23)/(df['ref_1_ND64']/0.13/0.13 - BG_1/0.23) 
df['T_1_150K'] = (df['Cu2O_1_150K_ND4']/0.23 - BG_1/0.23)/(df['ref_1_ND64']/0.13/0.13 - BG_1/0.23) 
df['T_1_296K'] = (df['Cu2O_1_296K_ND4']/0.23 - BG_1/0.23)/(df['ref_1_ND64']/0.13/0.13 - BG_1/0.23)
df
```

```python
plt.figure()
ax = df.plot('wl_1', 'T_1_296K', label=r'296 K')
df.plot('wl_1', 'T_1_150K', ax=ax, label=r'150 K')
df.plot('wl_1', 'T_1_77K', ax=ax, label=r'77 K')

#plt.xlim(400,700)
plt.xlim(560,660)
#plt.ylim(-0.02,0.12)
plt.ylim(0,0.13)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmittance')

plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure()
ax = df.plot('wl_2', 'ref_2_ND64')
df.plot('wl_2', 'Cu2O_2_77K_ND1', ax=ax)
df.plot('wl_2', 'BG_23_ND64', ax=ax)
plt.show()
```

```python
df['T_2_77K'] = (df['Cu2O_2_77K_ND1'] - BG_23/0.13/0.13)/(df['ref_2_ND64']/0.13/0.13 - BG_23/0.13/0.13)
```

```python
plt.figure()
ax = df.plot('wl_2', 'T_2_77K')
plt.show()
```

```python
plt.figure()
ax = df.plot('wl_3', 'ref_3_ND64')
df.plot('wl_3', 'Cu2O_3_77K_ND4', ax=ax)
df.plot('wl_3', 'BG_23_ND64', ax=ax)
plt.show()
```

```python
df['T_3_77K'] = (df['Cu2O_3_77K_ND4']/0.23 - BG_23/0.13/0.13)/(df['ref_3_ND64']/0.13/0.13 - BG_23/0.13/0.13)
```

```python
plt.figure()
ax = df.plot('wl_3', 'T_3_77K')
plt.show()
```

```python
df['OD_1_77K'] = - np.log10(df['T_1_77K'])
df['OD_1_150K'] = - np.log10(df['T_1_150K'])
df['OD_1_296K'] = - np.log10(df['T_1_296K'])
df.head()
```

```python
df['E_1'] = 1239.8/(df['wl_1'] + 1.7)
df['E_2'] = 1239.8/(df['wl_2'] + 1.7)
df['E_3'] = 1239.8/(df['wl_3'] + 1.7)
df
```

```python
plt.figure()
ax = df.plot('E_1', 'OD_1_296K', label=r'296 K')
df.plot('E_1', 'OD_1_150K', ax=ax, label=r'150 K')
df.plot('E_1', 'OD_1_77K', ax=ax, label=r'77 K')
plt.xlim(1.8, 2.4)
plt.ylim(0.80, 2.2)
plt.xlabel(r'Energy (eV)')
plt.ylabel(r'O.D.')
plt.grid()
plt.legend()
plt.show()
```
