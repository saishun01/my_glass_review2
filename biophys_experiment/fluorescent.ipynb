```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('fluorescent.csv')
df
```

```python
#GFPの各状態の蛍光スペクトル

plt.figure()
ax = df.plot('wl', 'int_nat', label='Native state')
df.plot('wl', 'int_unf', ax=ax, label='Unfolded state')
df.plot('wl', 'int_ref', ax=ax, label='Refolded state')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.xlim(360,760)
plt.ylim(0,1.2)
plt.legend()
plt.show()
```

```python
df.query('int_ref > 0.5')
```

## memo
Native state(int_nat) 509, 1.146981

Refolded state(int_ref) 509, 0.522644

```python
#励起光のスペクトル

plt.figure()
df.plot('wl', 'int_exc', legend=False)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.xlim(360,760)
plt.ylim(0,2500)
plt.show()
```

```python
df.query('int_exc > 2100')
```

```python
#タンパク質毎の蛍光スペクトル

plt.figure()
ax = df.plot('wl', 'int_nat', label='GFP')
df.plot('wl', 'int_yfp', ax=ax, label='YFP')
df.plot('wl', 'int_lys', ax=ax, label='Lysozyme')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.xlim(360,760)
plt.ylim(0,1.2)
plt.show()
```

```python
df.query('int_yfp > 0.25')
```
