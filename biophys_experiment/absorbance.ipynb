```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('absorbance.csv')
df
```

```python
plt.figure()
#ax = df.plot('wl', 'baseline')
ax = df.plot('wl', 'gfp', label='GFP')
df.plot('wl', 'yfp', ax=ax, label='YFP')
df.plot('wl', 'lysozyme', ax=ax, label='Lysozyme')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.xlim(350,550)
plt.ylim(0,0.3)
plt.legend()
plt.show()
```
