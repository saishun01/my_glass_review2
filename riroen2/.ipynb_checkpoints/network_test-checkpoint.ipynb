```python
import networkx as nx
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
G = nx.DiGraph()
nx.add_path(G, [3, 5, 4, 1, 0, 2, 7, 8, 9, 6])
nx.add_path(G, [3, 0, 6, 4, 2, 7, 1, 9, 8, 5])

nx.draw_networkx(G)
plt.show()
```

```python
#ノードの数
n = 7

#辺ができる確率
p = 0.3

#乱数のシード値
seed = 1

g = nx.random_graphs.fast_gnp_random_graph(n, p, seed, directed = True)
nx.draw_networkx(g)
plt.show()
```
